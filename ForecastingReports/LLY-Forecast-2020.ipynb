# Prediction Analysis of the Dataset

Fbprophet is a time-series data prediction analysis used to aid in understanding the trend and show its stock price depending on whether it is a year or more. This dataset is fed 5 years of data, however in the code above, it is set for a time period prediction for one year. 

Now, that we have imported the dataset and plotted the historical graph, we will predict its closing price one year.

```python
#Deploy the Fbprophet Prediction Library
import fbprophet
```

```python
l = fbprophet.Prophet()
```

```python
df = LLY
```

```python
df
```

```python
df[['ds','y']] = df[['Date','Close']]
```

```python
df = df[['ds','y']]
df.head()
```

```python
l.fit(df)
```

Below shows the future dates one year from the current stock market year.

```python
predict = l.make_future_dataframe(periods = 365)
predict.tail()
```

```python
predict = l.predict(predict)
predict.tail()
```

Prediction Output of the LLY (Eli Lilly and Company)

```python
l.plot(predict);
```

Below Shows the seasonality for trend, weekly, and yearly. These are essential when it comes to data analysis. 

Trend - according to the outputted graph below, we can see that it is predicting a closing price of $165.00/share. This prediction is shown assuming that the markets would rebound since the pandemic has caused havoc to the U.S. economy. Remember, this analysis is only a prediction based on how much data we fed to make a reasonable prediction.

Weekly - Furthermore, for the weekly analysis, we can see that for the day of Sunday; starts out at -0.5% loss, this trend shows that by Monday, the stock would be making a 0.4% gain and that gain keeps incrementing until Friday at 0.5%. This Gain would depend on factors, such as News, Business Decisions, .etc. The key part of analyzing seasonality is to better understand that the market can be making gains, but those gains could potentially go into a volatile trend. 

Yearly - Lastly, for the yearly analysis, there seems to be fluctuation in the stock's trend. We can see that this trend would decrease until an increase in the stock's value from November 1st of 2020 until it reaches its peak at January 1st of 2021 with a gain of 2.3%. Furthermore, this stock has potential investment in the long term.
