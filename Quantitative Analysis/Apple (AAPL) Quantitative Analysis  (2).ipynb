# Predicting Apple's (AAPL) Future Closing Price  

```python
a = fbprophet.Prophet()
```

```python
df = aapl
df
```

```python
df[['ds','y']]=df[['Date','Close']]
df=df[['ds','y']]
df
```

```python
a.fit(df)
```

```python
future = a.make_future_dataframe(periods=1825)
future.tail()
```

Below Illistrates the Dataset prediction in 5 years as well as its trend.

```python
future=a.predict(future)
future.tail()
```

# Prediction of the Dataset

Exponential Smoothing is also a time-series analysis, similar to Fbprophet, however, this method would be analyzed on the statistical side.

# Moving Average

Moving average is considered a simple technical analysis tool, much like exponential smoothing, it smooths out the datapoints by producing a constantly updated average price. This tool is aimed to help investors get vital data that determines what is the average stock price in any given time. It can be 10 days, 20 minutes, or weeks. 

Above, we can see that the moving average, similar to Exponential Smoothing, the data is smoothed out, this gives a better visualization of the data and it gives investors a better outlook about the average market value price. According to the Moving Average analysis, the average value is between $125.00-$292.00.
